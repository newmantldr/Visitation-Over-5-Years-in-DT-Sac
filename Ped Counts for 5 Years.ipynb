```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def loadFile(fileName, initial):
    
    f = open(os.path.expanduser("~/Desktop/current/five-year-visitation/"+fileName+"/daily.csv"))
    daily = pd.read_csv(f, header=0, names=['DOTW', 'PBID '+initial, 'OSW '+initial, 
                                               'CC '+initial, 'ED '+initial, 'K '+initial,
                                              'T '+initial, 'CM '+initial], index_col='DOTW')
    
    f = open(os.path.expanduser("~/Desktop/current/five-year-visitation/"+fileName+"/hourly.csv"))
    hourly = pd.read_csv(f, header=0, names=['Hour', 'PBID '+initial, 'OSW '+initial, 
                                               'CC '+initial, 'ED '+initial, 'K '+initial,
                                              'T '+initial, 'CM '+initial], index_col='Hour')
    
    f = open(os.path.expanduser("~/Desktop/current/five-year-visitation/"+fileName+"/visits.csv"))
    visits = pd.read_csv(f, header=0, names=['Visits', 'PBID', 'OSW', 'CC', 'ED', 'K', 'T', 'CM'], index_col='Visits')
    
    return daily, hourly, visits
```

```python
dailyOE, hourlyOE, visitsOE = loadFile('one-eight', 'OE')
dailyON, hourlyON, visitsON = loadFile('one-nine', 'ON')
dailyTZ, hourlyTZ, visitsTZ = loadFile('two-zero', 'TZ')
dailyTO, hourlyTO, visitsTO = loadFile('two-one', 'TO')
dailyTT, hourlyTT, visitsTT = loadFile('two-two', 'TT')
```

```python
dailyAllYearsList = [dailyOE, dailyON, dailyTZ, dailyTO, dailyTT]
hourlyAllYearsList = [hourlyOE, hourlyON, hourlyTZ, hourlyTO, hourlyTT]
visitsAllYearsList = [visitsOE, visitsON, visitsTZ, visitsTO, visitsTT]
```

```python
def concatDistricts(list1, list2, list3):
    
    dailyAllYears = pd.concat(list1, axis = 1)
    hourlyAllYears = pd.concat(list2, axis = 1)
    visitsAllYears = pd.concat(list3, axis = 0, keys = ['2017', '2018' '2019', '2020', '2021', '2022'])
    
    return dailyAllYears, hourlyAllYears, visitsAllYears
```

```python
dailyAllYears, hourlyAllYears, visitsAllYears = concatDistricts(dailyAllYearsList, 
                                                                hourlyAllYearsList, visitsAllYearsList)
```

```python
def plotDay(df, district):
    
    plt.figure(figsize=(16, 8), dpi=150)
    df[district+' OE'].plot(label = '2018')
    df[district+' ON'].plot(label = '2019')
    df[district+' TZ'].plot(label = '2020')
    df[district+' TO'].plot(label = '2021')
    df[district+' TT'].plot(label = '2022')
    plt.title('Visitation throughout a day in the '+district)
    plt.ylabel('Visits in Millions')
    plt.legend()
```
